```python
#Saurav Adhikari - 1622912
#Lalita Awasthi - 1622924
#Nila Ravindran - 1614113


#importing the required modules
from matplotlib import pyplot as plt
import numpy as np



x = np.linspace(-2., 2., 100)
y = np.linspace(-1., 3., 100)
X, Y = np.meshgrid(x,y)

#implementing the given rosenbrock function with a=1 ,b = 100
Z = (1-X) **2 + 100 * (Y-X**2 )**2

#plotting the contour
fig = plt.figure(0)
plt.contour(X, Y, Z, 30, colors = "black")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
