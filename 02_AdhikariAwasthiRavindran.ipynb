```python

#Saurav Adhikari - 1622912
#Lalita Awasthi - 1622924
#Nila Ravindran - 1614113

import numpy as np
from numdrv_test import test
from sympy import Symbol,diff

#function for central tendency
def drv(f,x,h):
    central_difference = (f(x+h) - f(x-h))/2*h
    return central_difference

#function for second derivative
def ddrv (f,x,h):
    second_derivative = (f(x+h) - 2 * f(x) + f(x-h))/h*h
    return second_derivative

#function for partial derivative
def pdrv(g,x,j,h):
    tmp = x[j]
    x[j] = h
    ret = g(x)
    x[j] = tmp
    return ret

#function for gradient
def grad(g,x,h):
    gradient = np.array(0)
    for j in range (0,2):
     np.append(gradient,pdrv(g,x,j,h))
    return gradient



#main method
if __name__ == '__main__':

    cent_ten = drv(np.cos,4,0.01)
    print(f'central tendency = {cent_ten}')
    test(ddrv)

    second_derivative = ddrv(np.sin,5,0.01)
    print(f'second_derivative = {second_derivative}' )
    test(ddrv)


    print("Test for partial derivative")
    test(pdrv)

    print("Test for gradient")
    test(grad)
```
